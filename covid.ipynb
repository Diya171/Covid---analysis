```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('country_wise_latest.csv.xlsx')
```

```python
df.head()
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
print(df.isnull().sum())
```

```python
df["Deaths / 100 Recovered"].fillna(df["Deaths / 100 Recovered"].mean(), inplace=True)
```

```python
global_totals = df[["Confirmed","Deaths","Recovered","Active"]].sum()
print(global_totals)
```

```python
top50_cases = df.sort_values(by="Confirmed", ascending=False).head(50)
print(top50_cases[["Country/Region","Confirmed"]])
```

```python
top50_deaths = df.sort_values(by="Deaths", ascending=False).head(50)
print(top50_deaths[["Country/Region","Deaths"]])
```

```python
top50_recovered = df.sort_values(by="Recovered", ascending=False).head(50)
print(top50_recovered[["Country/Region","Recovered"]])
```

```python
top50_deaths_after_recover_per100 = df.sort_values(by="Deaths / 100 Recovered", ascending=False).head(50)
print(top50_deaths_after_recover_per100[["Country/Region","Deaths / 100 Recovered"]])
```

```python
plt.figure(figsize=(15,7))
plt.hist(df["Confirmed"], bins=50, color="yellow", edgecolor="black", log=True) # Added log=True for logarithmic scale
plt.xlabel("Confirmed Cases (Log Scale)") # Updated label
plt.ylabel("Frequency")
plt.title("Distribution of Confirmed Cases (Log Scale)") # Updated title
plt.show()
```

```python
plt.figure(figsize=(15, 7))
plt.bar(top50_cases["Country/Region"], top50_cases["Confirmed"], color="skyblue", align="edge", width=0.8, edgecolor="black")
plt.xlabel("Country/Region")
plt.ylabel("Confirmed Cases")
plt.title("Top 50 Countries with Highest Confirmed Cases")
plt.xticks(rotation=90)
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```

```python
plt.figure(figsize=(15, 7))
plt.bar(top50_deaths["Country/Region"], top50_deaths["Confirmed"], color="red", align="edge", width=0.8, edgecolor="black")
plt.xlabel("Country/Region")
plt.ylabel("Total Deaths")
plt.title("Top 50 Countries with Highest Deaths")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(15, 7))
plt.plot(top50_recovered["Country/Region"], top50_recovered["Recovered"], color="black", lw=0.8, marker = "o", markersize = 3)
plt.xlabel("Country/Region")
plt.ylabel("Recovered")
plt.title("Top 50 Countries with Highest Recovered")
plt.xticks(rotation=90)
plt.tight_layout() # Adjust layout to prevent labels overlapping
plt.show()
```

```python
plt.figure(figsize=(15, 7))
plt.bar(top50_deaths_after_recover_per100["Country/Region"], top50_deaths_after_recover_per100["Deaths / 100 Recovered"], color="green", align="edge", width=0.8, edgecolor="black", log=True)
plt.xlabel("Country/Region")
plt.ylabel("Total Deaths after recovery")
plt.title("Top 50 Countries with Highest Deaths after recovery")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
explodes=[0,0.1,0,0]
plt.figure(figsize=(6,6))
plt.pie(global_totals, labels=global_totals.index, explode=explodes, autopct="%1.1f%%", pctdistance=0.8, startangle=90)
plt.title("COVID-19 Global Distribution")
plt.show()
```

```python
print("Mean cases:", np.mean(df["Confirmed"]))
print("Median cases:", np.median(df["Confirmed"]))
print("Std dev cases:", np.std(df["Confirmed"]))
print("Correlation:\n", df[["Confirmed","Deaths"]].corr())
```

```python
region_summary = df.groupby("WHO Region")[["Confirmed","Deaths","Recovered","Active"]].sum()
print(region_summary)

# Plot
region_summary.plot(kind="bar", figsize=(12,6))
plt.title("COVID-19 Cases by WHO Region")
plt.ylabel("Counts")
plt.show()
```
